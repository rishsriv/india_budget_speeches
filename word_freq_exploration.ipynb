### A basic analysis of word frequencies in Indian budget speeches

Setting up imports

```python
%matplotlib inline

import pandas as pd
import os
import string
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

Preparing the data

```python
filenames = []
for fname in os.listdir('txt'):
    if fname.endswith('.txt'):
        filenames.append(fname)
filenames = sorted(filenames)
```

```python
d = {}
for fname in filenames:
    with open('txt/%s'%fname, 'r') as f:
        s = str(f.read().decode('ascii', 'ignore').lower())
    s = [i for i in s.translate(string.maketrans("",""), string.punctuation).split() if not i.isdigit()]
    clean_name = lambda x: x[:4] + '-' + x[4:]
    d[clean_name(fname.split('.')[0])] = Counter(s)

wt = pd.Series({y: sum(d[y].values()) for y in d})
```

Plotting the data

```python
def plot_chart(words):
    wc = pd.Series({y: sum([d[y][word] for word in words]) for y in d})
    wf = 100.*wc/wt
    wc.plot(kind='area', figsize=(12,6))
    #wf.plot(kind='area', figsize=(12,6))
    plt.title(words)
    wc.to_csv('data/%s.csv'%'-'.join(words))
    plt.ylabel('Number of times used')
    plt.savefig('data/%s.png'%'-'.join(words))
```

```python
wt.plot(kind='area', title='Number of words in Budget Speech', figsize=(12,6))
wt.to_csv('data/total.csv')
plt.ylabel('Number of times used')
plt.savefig('data/%s.png'%'total')
```
